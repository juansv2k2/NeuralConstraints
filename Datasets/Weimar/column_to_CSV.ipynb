```python
import pandas as pd
```

```python
df = pd.read_csv('./folk/melody.csv')
df.head(10)
```

```python
# create a mask
mask = df['melid'] <= 1000

# apply the mask to select rows where melid is less than 1000
pitch_column = df.loc[mask, 'pitch']

pitch_column_int = pitch_column.apply(lambda x: int(x))

pitch_column_int.shape
```

```python
pitch_column_int.to_csv('pitch_column.txt', header=None, index=None, sep=' ')
```
