```python
import pandas as pd
from simplefractions import *
```

```python
df = pd.read_csv('../../../datasets/folk_melody.csv')
df.head(10)
```

```python
# create a mask
mask = df['melid'] <= 100

# apply the mask to select rows where melid is less than 1000

pitch_duration_columns = df.loc[mask, ['pitch','duration','num']]


pitch_duration_columns['pitch'] = pitch_duration_columns['pitch'].apply(lambda x: int(x))
pitch_duration_columns['duration'] = pitch_duration_columns.apply(lambda row: row['duration'] / row['num'], axis=1)
pitch_duration_columns['duration'] = pitch_duration_columns['duration'].apply(lambda x: simplest_from_float(x))
pitch_duration_columns = pitch_duration_columns.drop('num', axis=1)


pitch_duration_columns.head(20)

# pitch_column_int.shape
```

```python
pitch_duration_columns.to_csv('pitch_<100.txt', header=None, index=None, sep=' ')
```
